```python
def calc_output(W1, W2, x):
    v1 = np.matmul(W1, x)
    y1 = sigmoid(v1)
    v = np.matmul(W2, y1)
    y = sigmoid(v)

    return y, y1
```

```python
def calc_delta(d, y):
    e = d - y
    delta = y * (1 - y) * e

    return delta
```

```python
def calc_delta1(W2, delta, y1):
    e1 = np.matmul(W2.T, delta)
    delta1 = y1 * (1 - y1) * e1

    return delta1
```
